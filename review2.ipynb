```python
import pandas as pd
d=pd.read_csv('/content/box office prediction 777.csv')
print(d)
```

```python
x=d.iloc[:,1:8]
y=d.iloc[:,8:9]
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
print(x.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=True,test_size=0.20)
```

```python
print(x_train,y_train)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
mm=lr.fit(x_train,y_train)
```

```python
yp=mm.predict(x_test)
```

```python
print(yp)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(yp,y_test))
```

```python
from sklearn.metrics import mean_absolute_error
print(mean_absolute_error(yp,y_test))
```

```python
from sklearn.neighbors import KNeighborsRegressor
model=KNeighborsRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_pred)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
```

```python
from sklearn.svm import SVR
model=SVR()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_pred)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(y_pred)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred))
from sklearn import tree
tree.plot_tree(model,filled=True)
```
